# Descomprimir Dataset

```python
# Rutas
zip_real = '/content/drive/MyDrive/Proyecto_Grado/Data/frames_extraidos.zip'
extract_real = '/content/drive/MyDrive/Proyecto_Grado/Data/frames_extraidos'

# Crear carpeta destino
os.makedirs(extract_real, exist_ok=True)

# Extraer ignorando carpetas internas
with zipfile.ZipFile(zip_real, 'r') as zip_ref:
    for member in zip_ref.namelist():
        filename = os.path.basename(member)
        if filename:  # Evita carpetas vacías
            source = zip_ref.open(member)
            target = open(os.path.join(extract_real, filename), "wb")
            with source, target:
                target.write(source.read())

# Listar imágenes extraídas
image_paths = sorted([
    str(p) for p in pathlib.Path(extract_real).glob('*')
    if p.suffix.lower() in ['.jpg', '.jpeg', '.png']
])

# Contar
print(f"Total de imágenes extraídas: {len(image_paths)}")

# Mostrar 5 imágenes aleatorias
if len(image_paths) > 0:
    sample_paths = random.sample(image_paths, min(5, len(image_paths)))
    plt.figure(figsize=(15, 5))
    for i, img_path in enumerate(sample_paths):
        img = Image.open(img_path).convert("RGB")
        plt.subplot(1, 5, i+1)
        plt.imshow(img)
        plt.title(f"Imagen {i+1}")
        plt.axis("off")
    plt.show()
else:
    print("⚠ No se encontraron imágenes.")
```
